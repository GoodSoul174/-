## Описание проекта

Необходимо разработать **алгоритм прогнозирования спроса и пользовательский интерфейс для визуализации прогнозов** по товарам собственного производства (СП). Цель проекта – **оптимизировать производство товаров СП**, повысить их доступность в магазинах и увеличить продажи **без роста списаний**, что в свою очередь повысит маржинальность.

Ранее сотрудники производства опирались на экспертные оценки, что приводило к неточностям в заказах сырья и выпуске продукции. Внедрение **ML-прогноза спроса** позволит автоматизировать этот процесс, используя данные о продажах, промо-активности, характеристиках товаров и сезонности.

### Цели сервиса

**Модели Data Science (DS):**

✅ DS: Прогнозирование спроса на 14 дней с ежедневным обновлением на основе данных о продажах, товарах, магазинах и календарных признаках.

**Функциональность интерфейса (Tool):**

✅ Визуализация прогноза спроса на товары СП в разрезе ТК-SKU-День.

✅ Оценка точности модели – сравнение предсказаний с фактическими продажами на исторических данных (в таблицах и графиках).

✅ Гибкость анализа – возможность просмотра данных в различных детализациях (от сети магазинов до конкретного ТК).

````python
# Настройка стиля графиков
sns.set_style("whitegrid")

# Функция для создания скользящей статистики
def plot_rolling_stats(df, title):
    fig, ax = plt.subplots(figsize=(14, 7))
    
    # Расчет статистик
    df['mean_30'] = df['pr_sales_in_rub'].rolling(30).mean()
    df['std_30'] = df['pr_sales_in_rub'].rolling(30).std()
    df['upper'] = df['mean_30'] + df['std_30']
    df['lower'] = df['mean_30'] - df['std_30']
    
    # Визуализация
    df['pr_sales_in_rub'].plot(ax=ax, color='#1f77b4', alpha=0.5, label='Дневные продажи')
    df['mean_30'].plot(ax=ax, color='#2ca02c', linewidth=2, label='Скользящее среднее (30 дней)')
    ax.fill_between(df.index, df['lower'], df['upper'], color='#ff7f0e', alpha=0.1, label='±1σ')
    
    # Разметка праздников
    holidays = {
        'Новый год': '2022-12-31',
        '8 Марта': '2023-03-08',
        'Майские': '2023-05-01'
    }
    
    for name, date in holidays.items():
        ax.axvline(pd.to_datetime(date), color='red', linestyle='--', alpha=0.5)
        ax.text(pd.to_datetime(date), ax.get_ylim()[1]*0.9, name, rotation=90, ha='right')
    
    # Настройки графика
    ax.set_title(f'Анализ стационарности: {title}', pad=20, fontsize=14)
    ax.set_ylabel('Сумма продаж, руб', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend(loc='upper left', fontsize=10)
    plt.tight_layout()
    plt.show()

# Анализ для обычных продаж
plot_rolling_stats(df_sales_rub.resample('1D').sum(), 'обычные продажи')

# Анализ для промо-продаж
plot_rolling_stats(df_sales_rub_promo.resample('1D').sum(), 'промо-продажи')
````

````python
# 2. Декомпозиция временных рядов
df_resample = train_rub.resample('1D').sum()
df_resample_promo = train_rub_promo.resample('1D').sum()

# Функция для декомпозиции и визуализации
def plot_decomposition(data, title, period=7):
    decomposed = seasonal_decompose(data, period=period)
    
    plt.figure(figsize=(12, 10))
    
    # Тренд
    plt.subplot(311)
    decomposed.trend.plot()
    plt.title('Тренд')
    plt.grid(True)
    
    # Сезонность (2 недели вокруг Нового года)
    plt.subplot(312)
    decomposed.seasonal['2022-12-15':'2023-01-01'].plot()
    plt.title('Сезонность на 2 недели')
    plt.grid(True)
    
    # Остатки
    plt.subplot(313)
    decomposed.resid.plot()
    plt.title('Остатки')
    plt.grid(True)
    
    plt.suptitle(f'Декомпозиция {title}', y=1.02, fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Рассчитываем и выводим метрики
    trend_series = decomposed.trend.dropna()
    if len(trend_series) > 0:
        trend_growth = (trend_series.iloc[-1] - trend_series.iloc[0]) / trend_series.iloc[0]
    else:
        trend_growth = 0
    
    print(f"\nКлючевые метрики ({title}):")
    print(f"Амплитуда сезонности: {decomposed.seasonal.max():.0f} единиц")
    print(f"Общий рост тренда: {trend_growth:.2%} за период")
    print(f"Стандартное отклонение остатков: {decomposed.resid.std():.2f}")

# Анализ для продаж без промо
plot_decomposition(df_resample, 'продаж без промо')

# Анализ для продаж с промо
plot_decomposition(df_resample_promo, 'продаж с промо')
````

````python
# 3. Скользящие статистики
def plot_rolling_stats(data, title):
    stats = data.resample('1D').sum()
    rolling = stats.rolling(30)
    
    plt.figure(figsize=(12, 6))
    plt.plot(stats.index, stats['pr_sales_in_units'], label='Факт', alpha=0.5)
    
    # Правильный расчет std и mean
    mean = rolling.mean()['pr_sales_in_units']
    std = rolling.std()['pr_sales_in_units']
    
    plt.plot(stats.index, mean, label='Средняя', color='green')
    plt.fill_between(stats.index, 
                    mean - std, 
                    mean + std,
                    color='gray', alpha=0.2, label='±1σ')
    
    plt.title(f'{title} (скользящее окно 30 дней)')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_rolling_stats(train_rub, 'Продажи без промо')
plot_rolling_stats(train_rub_promo, 'Продажи с промо')
````

### Расчет цены за единицу товара

**Метрика WAPE (Weighted Absolute Percentage Error)**  
- **Что измеряет?**  
  Средневзвешенную ошибку прогноза в процентах от общего объема продаж.  
  Формула: `WAPE = Σ|Факт − Прогноз| / Σ|Факт|`.  

- **Почему выбрана?**  
  - Устойчива к разномасштабным данным (например, товары с ценами 100 руб. и 10 000 руб. учитываются пропорционально).  
  - Лучше MAPE при нулевых продажах (не дает бесконечности).  
  - Интерпретируема: WAPE = 0.15 означает ошибку 15% от общего оборота.  

- **Как интерпретировать?**  
  Чем ближе к 0, тем точнее модель. Пример:  
  - WAPE=0.1 → хороший прогноз (ошибка 10%).  
  - WAPE=0.5 → неприемлемо (ошибка 50%).  

````python
# Создаем копию DataFrame, если train - это срез другого DataFrame
train = train.copy()

# Расчет цены за единицу товара с защитой от деления на ноль
train.loc[:, 'pr_sales_in_rub_id'] = round(
    train['pr_sales_in_rub'] / train['pr_sales_in_units'].replace(0, 1),
    1
)

# Заполнение пропусков (если появятся)
train.loc[:, 'pr_sales_in_rub_id'] = train['pr_sales_in_rub_id'].fillna(0)
````

  Признак `pr_sales_in_rub` (общая выручка) напрямую зависит от количества проданных единиц (`pr_sales_in_units`), что приводит к:  
  - **Мультиколлинеарности**: Модель "жульничает", используя линейную связь между суммой и количеством.  
  - **Переобучению**: Искусственно завышенная точность на тесте, но плохие прогнозы на новых данных.  

- **Решение:**  
  Замена на **цену за штуку** (`pr_sales_in_rub_id` = `pr_sales_in_rub / pr_sales_in_units`).  

- **Преимущества:**  
  - Убирает зависимость между признаками.  
  - Лучше отражает реальную стоимость товара (например, цена молока = 80 руб./шт., а не 8000 руб. за 100 шт.).  

- **Обработка крайних случаев:**  
  - Если товар не продан (`pr_sales_in_units=0`), цена = 0 (бесплатный товар) или заменяется на среднюю по категории.  
  - Выбросы (например, цена 1 млн руб. за штуку) фильтруются.

````python
def make_features(df_mf, max_lag, rolling_mean_size):
    """
    Создает временные признаки, лаги и скользящие средние для временных рядов.
    Особенности:
    - Циклическое кодирование времени (sin/cos)
    - Лаги продаж от 14 до max_lag дней
    - Скользящее среднее с окном rolling_mean_size
    - Оптимизированное заполнение пропусков
    """
    df_mf = df_mf.copy()
    df_mf_full_list = []
    
    # Циклические признаки времени
    df_mf['day'] = df_mf.index.day
    df_mf['dayofweek'] = df_mf.index.dayofweek
    df_mf['month'] = df_mf.index.month
    
    # Нормализованные циклические признаки
    for period, max_val in [('dayofweek', 7), ('month', 12), ('day', 31)]:
        df_mf[f'sin_{period}'] = np.sin(2 * np.pi * df_mf[period]/max_val)
        df_mf[f'cos_{period}'] = np.cos(2 * np.pi * df_mf[period]/max_val)
    
    # Создаем лаги и скользящие средние для каждого типа продаж
    for s in df_mf['pr_sales_type_id'].unique():
        df_mf_temp = df_mf.loc[df_mf['pr_sales_type_id'] == s].copy()
        
        # Лаги от 14 до max_lag дней
        lags = range(14, max_lag + 1)
        for lag in lags:
            df_mf_temp[f'sales_lag_{lag}'] = df_mf_temp.groupby(
                ["st_id", "pr_sku_id"])['pr_sales_in_units'].shift(lag)
        
        # Скользящее среднее с центрированием
        df_mf_temp['rolling_mean_units'] = (
            df_mf_temp.groupby(["st_id", "pr_sku_id"])['pr_sales_in_units']
            .transform(lambda x: x.shift(14).rolling(rolling_mean_size, min_periods=1).mean())
        )
        
        df_mf_full_list.append(df_mf_temp)
    
    # Объединяем все типы продаж
    df_mf_full = pd.concat(df_mf_full_list)
    
    # Удаляем временные колонки
    time_cols = ['day', 'dayofweek', 'month']
    df_mf_full = df_mf_full.drop(columns=time_cols)
    
    # Заполняем пропуски в лагах и средних
    lag_cols = [col for col in df_mf_full.columns if 'lag_' in col or 'rolling_' in col]
    df_mf_full[lag_cols] = df_mf_full[lag_cols].fillna(0)
    
    # Для новых товаров устанавливаем флаг
    df_mf_full['is_new_product'] = df_mf_full.groupby(
        ['st_id', 'pr_sku_id'])['pr_sales_in_units'].transform(lambda x: x.notna().sum() == 0).astype(int)
    
    return df_mf_full

# Применяем функцию
df_mf_full = make_features(combined, 28, 14)

# Проверка результатов
print("Статистика по пропускам после заполнения:")
print(df_mf_full.isna().sum().sort_values(ascending=False).head(10))
print("\nИнформация о датасете:")
df_mf_full.info()
````

````python
# Временное разделение внутри df_mf (не затрагивая test_mf!)
train_p = df_mf['2022-08-01':'2023-07-04']
test_p = df_mf['2023-07-04':'2023-07-18']

# Подготовка данных для Prophet
def prepare_for_prophet(df):
    df = df.reset_index()[['date', 'st_id', 'pr_sku_id', 'pr_sales_in_units', 'pr_sales_in_rub_id', 'pr_sales_type_id']]
    df = df.rename(columns={'date': 'ds', 'pr_sales_in_units': 'y'})
    df['st_pr_sku'] = df['st_id'] + '_' + df['pr_sku_id']
    return df

pr_train = prepare_for_prophet(train_p)
pr_test = prepare_for_prophet(test_p)

# 1. Создаем календарь праздников РФ
ru_holidays = make_holidays_df(year_list=[2022, 2023], country='RU')

# 2. Добавляем регрессоры
pr_train['price'] = pr_train['pr_sales_in_rub_id']
pr_train['is_promo'] = (pr_train['pr_sales_type_id'] == 1).astype(int)

pr_test['price'] = pr_test['pr_sales_in_rub_id']
pr_test['is_promo'] = (pr_test['pr_sales_type_id'] == 1).astype(int)

# 3. Обучение Prophet

models = {}
for id in pr_train['st_pr_sku'].unique():
    train_ = pr_train[pr_train['st_pr_sku'] == id]
   
    if len(train_) > 2:
        m = Prophet(
            seasonality_mode='multiplicative',
            yearly_seasonality=True,
            weekly_seasonality=True,
            daily_seasonality=False,
            changepoint_prior_scale=0.01,
            holidays=ru_holidays
        )
        m.add_regressor('price')
        m.add_regressor('is_promo')
        m.fit(train_)
        models[id] = m
````

````python
# 4. Прогнозирование и оценка WAPE
forecasts = []
for id, model in models.items():
    future = model.make_future_dataframe(periods=14)
    train_ = pr_train[pr_train['st_pr_sku'] == id]
    test_subset = pr_test[pr_test['st_pr_sku'] == id]
    future = future.merge(test_subset[['ds', 'price', 'is_promo']], on='ds', how='left')
    future['price'] = future['price'].fillna(train_['price'].median())
    future['is_promo'] = future['is_promo'].fillna(0)
    
    forecast = model.predict(future)
    forecast['st_pr_sku'] = id
    forecasts.append(forecast[['ds', 'st_pr_sku', 'trend', 'weekly', 'yhat', 'is_promo']])
    
forecast['is_promo'] = forecast['is_promo'].astype(int)
p = pd.concat(forecasts)
val_results = pr_test.merge(p, on=['ds', 'st_pr_sku', 'is_promo'], how='left')

# Фильтрация и расчет WAPE
val_results = val_results[val_results['y'] > 0]  # Игнорируем нулевые продажи
val_results['yhat'] = val_results['yhat'].clip(lower=0)  # Отрицательные -> 0

wape = np.sum(np.abs(val_results['y'] - val_results['yhat'])) / np.sum(val_results['y'])
print(f"WAPE Prophet с праздниками и регрессорами: {wape:.1%}")
````

````python
def add_prophet_features(df, models, is_train=True):
    """
    Добавляет признаки Prophet к данным.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Данные (трейн или тест).
    models : dict
        Словарь моделей Prophet, где ключи — st_pr_sku.
    is_train : bool
        Флаг: True для трейна, False для теста.
    """
    forecasts = []
    
    for st_pr_sku, model in models.items():
        # Создаем future dataframe
        if is_train:
            future = model.make_future_dataframe(periods=0)
        else:
            # Для теста берем уникальные даты и указываем periods
            test_dates = df['ds'].nunique()
            future = model.make_future_dataframe(
                periods=test_dates,
                include_history=False
            )
        
        # Добавляем price и is_promo
        subset = df[df['st_pr_sku'] == st_pr_sku]
        future = future.merge(
            subset[['ds', 'price', 'is_promo']],
            on='ds',
            how='left'
        )
        
        future['price'] = future['price'].fillna(df['price'].median())
        future['is_promo'] = future['is_promo'].fillna(0)
        
        # Прогнозируем
        forecast = model.predict(future)
        forecast['is_promo'] = future['is_promo'].values
        forecast['st_pr_sku'] = st_pr_sku
        forecasts.append(forecast[['ds', 'trend', 'weekly', 'yhat', 'st_pr_sku', 'is_promo']])
        
    # Объединяем прогнозы
    prophet_features = pd.concat(forecasts).rename(columns={'ds': 'date'}).drop_duplicates(['date', 'st_pr_sku', 'is_promo'])
    
    # Мержим с исходными данными
    if 'st_pr_sku' not in df.columns:
        df['st_pr_sku'] = df['st_id'] + '_' + df['pr_sku_id']

    if 'ds' in df.columns:
        df = df.rename(columns={'ds': 'date'})
    
    return df[['date', 'st_pr_sku', 'is_promo']].merge(prophet_features, on=['date', 'st_pr_sku', 'is_promo'], how='left')

# Добавляем признаки к трейну и тесту
df_mf = df_mf.merge(add_prophet_features(mf_train, final_models, is_train=True), on=['date', 'st_pr_sku', 'is_promo'], how='left')
test_mf = test_mf.merge(add_prophet_features(mf_test, final_models, is_train=False), on=['date', 'st_pr_sku', 'is_promo'], how='left')

# Создаем маску на валидационные даты

val_mask = df_mf['date'].isin(val_results['ds'])

# Выбираем только валидационные строки
df_val = df_mf.loc[val_mask].copy()

# Мержим val_results, чтобы взять корректные значения признаков
df_val_updated = df_val.merge(
    val_results[['ds', 'st_pr_sku', 'is_promo', 'yhat', 'trend', 'weekly']],
    left_on=['date', 'st_pr_sku', 'is_promo'],
    right_on=['ds', 'st_pr_sku', 'is_promo'],
    how='left',
    suffixes=('', '_val')
)

assert len(df_val) == len(df_val_updated), "Размерности после merge не совпадают!"

# Сохраняем индекс, чтобы можно было присваивать
df_val_updated.set_index(df_val.index, inplace=True)

for col in ['yhat', 'trend', 'weekly']:
    df_mf.loc[val_mask, col] = df_val_updated[f"{col}_val"]

print("\nОбновленные признаки:")
print(df_mf[['date', 'st_id', 'pr_sku_id', 'trend', 'weekly', 'yhat']].head())
````

### Поиск лучшей модели

````python
# --- Метрики ---
def wape(y_true: np.array, y_pred: np.array) -> float:
    denominator = np.sum(np.abs(y_true))
    return np.sum(np.abs(y_true - y_pred)) / denominator if denominator != 0 else np.nan

wape_scorer = make_scorer(wape, greater_is_better=False)

# --- Подготовка данных ---
features_cat = X_train.select_dtypes(include='object').columns.to_list()

# --- Target (mean) encoding ---
def mean_encode(X, y, features_cat):
    X = X.copy()
    for col in features_cat:
        means = y.groupby(X[col]).mean()
        X[col] = X[col].map(means).fillna(0)
    return X

def mean_encoding_wrapper(X):
    return mean_encode(X, y_train, features_cat)

# --- Категориальные фичи для CatBoost ---
cat_features_index = [X_train.columns.get_loc(col) for col in features_cat]

# --- Pipeline с разными кодировками ---
# 1. Ordinal
ordinal_encoder = ColumnTransformer([
    ("ord", OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), features_cat)
], remainder='passthrough')

# 2. Mean Encoding
mean_encoder = FunctionTransformer(func=mean_encoding_wrapper, validate=False)

# --- Обучающие выборки для разных моделей ---
# Ordinal
train_ord = pd.DataFrame(ordinal_encoder.fit_transform(X_train), columns=X_train.columns)
val_ord = pd.DataFrame(ordinal_encoder.transform(X_val), columns=X_train.columns)
train_ord_target = y_train
val_ord_target = y_val

# Mean Encoding
train_mean = mean_encoding_wrapper(X_train)
val_mean = mean_encoding_wrapper(X_val)
train_mean_target = y_train
val_mean_target = y_val

# --- Модели ---
models = {
    'RandomForest_Ordinal': Pipeline([
        ("encoder", ordinal_encoder),
        ("model", RandomForestRegressor(random_state=42))
    ]),
    'RandomForest_MeanEnc': Pipeline([
        ("encoder", mean_encoder),
        ("model", RandomForestRegressor(random_state=42))
    ]),
    'LGBM_Ordinal': Pipeline([
        ("encoder", ordinal_encoder),
        ("model", LGBMRegressor(random_state=42))
    ]),
    'LGBM_MeanEnc': Pipeline([
        ("encoder", mean_encoder),
        ("model", LGBMRegressor(random_state=42))
    ]),
    'CatBoost_Native': CatBoostRegressor(cat_features=cat_features_index, random_state=42, verbose=0)
}

# --- Обучение и метрики ---
results = {}

for name, model in models.items():
    if 'Ordinal' in name:
        model.fit(train_ord, train_ord_target)
        y_pred = model.predict(val_ord)
        y_true = val_ord_target

    elif 'MeanEnc' in name:
        model.fit(train_mean, train_mean_target)
        y_pred = model.predict(val_mean)
        y_true = val_mean_target

    elif 'CatBoost' in name:
        model.fit(X_train, y_train, cat_features=cat_features_index)
        y_pred = model.predict(X_val)
        y_true = y_val

    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    wape_score = wape(y_true, y_pred)

    results[name] = {'RMSE': rmse, 'WAPE': wape_score}

# --- Вывод ---
print(f"{'Model':<25} {'RMSE':<10} {'WAPE':<10}")
print("-" * 45)
for name, metrics in results.items():
    print(f"{name:<25} {metrics['RMSE']:<10.2f} {metrics['WAPE']:<10.4f}")

print(f"MAE: {mean_absolute_error(y_val, y_pred):.2f}")
````
